```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

from libs import utils             ## customs built functions libries, some basic image minipulation functions
```

### load images  -- they are all the same size

```python
## load 50 images into a 4d numpy array, each row is a 3d image
files = utils.get_celeb_files(dst='../data/0_img_align_celeba',max_images=150)  ## grab 50 images 
imgs=[plt.imread(f) for f in files]
data = np.array(imgs)
data.shape
```

### Mean/Deviation of Images

```python
## mean of the batch - average all accross all imgs 
## omg!!!! 
mean_img = np.mean(data,axis=0)
plt.imshow(mean_img.astype(np.uint8))    ## astype np.unit8 turns all floating number to integers 
```

```python
## look at the standard deviation 
std_img = np.std(data,axis=0)
plt.imshow(std_img.astype(np.uint8))
## it is showing the variance of previous mean image
```

```python
## show the mean of the variance 
plt.imshow(np.mean(std_img,axis=2).astype(np.uint8))
```
